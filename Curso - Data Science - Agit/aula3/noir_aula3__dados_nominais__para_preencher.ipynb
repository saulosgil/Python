# Escala Nominal


Uma escala nominal designa nmeros para rotular ou identificar indivduos ou objetos.

Dados nominais representam somente categorias ou classes, e no implicam quantias de um atributo ou caracterstica. 

Exemplos freqentemente usados de dados nominalmente escalonados incluem: 

1. **atributos demogrficos** <font color='grey'>: sexo, religio, ocupao ou filiao partidria</font>
2. **muitas formas de comportamento**: <font color='grey'>comportamento de voto ou atividade de compra</font>
3. **qualquer outra ao que seja discreta**: <font color='grey'>que acontece ou no</font>.

**Caractersticas da escala nominal:**

1. Qualitativa / Categrica
2. No existe ordem entre as categorias: Ou so iguais, Ou diferentes
3. Medies
    1. Qui Quadrado
    2. Moda (o que mais aparece)
    3. Percentual Modal 
    4. Distribuio de Frequncia
    6. No se pode calcular a mdias
4. Grficos
    1. Pizza
    2. Barras
