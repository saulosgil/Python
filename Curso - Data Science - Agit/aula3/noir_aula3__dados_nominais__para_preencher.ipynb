# Escala Nominal


Uma escala nominal designa números para rotular ou identificar indivíduos ou objetos.

Dados nominais representam somente categorias ou classes, e não implicam quantias de um atributo ou característica. 

Exemplos freqüentemente usados de dados nominalmente escalonados incluem: 

1. **atributos demográficos** <font color='grey'>: sexo, religião, ocupação ou filiação partidária</font>
2. **muitas formas de comportamento**: <font color='grey'>comportamento de voto ou atividade de compra</font>
3. **qualquer outra ação que seja discreta**: <font color='grey'>que acontece ou não</font>.

**Características da escala nominal:**

1. Qualitativa / Categórica
2. Não existe ordem entre as categorias: Ou são iguais, Ou diferentes
3. Medições
    1. Qui Quadrado
    2. Moda (o que mais aparece)
    3. Percentual Modal 
    4. Distribuição de Frequência
    6. Não se pode calcular a médias
4. Gráficos
    1. Pizza
    2. Barras
