# Conversão entre Tipos

Muitas vezes é necessário converter um tipo de dados em outro.

Basicamente há 2 casos:
 1. Números que precisam se tornar textos
 2. Textos em que precisamos extrair números


|De|Para|Exemplo - Entrada|Exemplo - Saída|
|--|----|-------|----|
|int|str|`str(10)`|'10'|
|float|str|`str(76.5)`|'76.5'|
|str|int|`int('1984')`|1984|
|str|float|`float('3.1416')`|3.1416|

```python
def calcular_tempo(horas, resposta_em_minutos = False):

    unidade = 's'
    resposta = 3600 * horas
    ################################################
    if resposta_em_minutos:   
        resposta /= 60 # resposta = resposta / 60 
        unidade = 'min'
    ################################################
    return f'{horas}h tem {resposta:,}{unidade}'
```

## Múltiplas condições

Uso do `elif` e do `else`

```python
def obter_dia_semana_texto(dia_semana_num):    
    if dia_semana_num == 1:
        dia_semana_nome = 'domingo'
    elif dia_semana_num == 2:
        dia_semana_nome = 'segunda'
    elif  dia_semana_num == 3:
        dia_semana_nome = 'terça'
    elif dia_semana_num == 4:
        dia_semana_nome = 'quarta'
    elif dia_semana_num == 5:
        dia_semana_nome = 'quinta'
    elif dia_semana_num == 6:
        dia_semana_nome = 'sexta'
    elif dia_semana_num == 7:
        dia_semana_nome = 'sábado'
    else:
        dia_semana_nome = 'Valor inválido: entre com um número entre 1 e 7'
    return dia_semana_nome    
```

b) Altere a função para que:
  - Receba um **novo parâmetro booleano** chamado `em_minutos`
  - Transforme o valor de dias para:
    - **minutos**, se o novo parâmetro for igual a `True`
    - **horas**, se o novo parâmetro for igual a `False`

```python
def calcular_tempo(dias):
    return 24 * dias
```

```python
def calcular_tempo(dias, em_minutos):
    if em_minutos:
        resultado = dias * 24 * 60
    else:
        resultado = dias * 24
    return resultado
```

c) Altere a função novamente para que:

  - Receba um **novo parâmetro string** chamado `unidade`
  - Transforme o valor de dias para:
    - **segundos**, se o novo parâmetro for igual a `s`
    - **minutos**, se o novo parâmetro for igual a `min`
    - **horas**, se o novo parâmetro for igual a  `h`

```python
def calcular_tempo(dias):
    return 24 * dias
```

```python
def calcular_tempo(dias, unidade):
    if unidade == 's':
        resultado = 24 * 60 * 60 * dias
    elif unidade == 'min':
        resultado = 24 * 60 * dias
    elif unidade == 'h':
        resultado = 24 * dias
    return resultado
```

```python
assert calcular_tempo(2, 's') == 172800
```

```python
assert calcular_tempo(2, 'min') == 2880
```

```python
assert calcular_tempo(2, 'h') == 48
```

Estruturas de dados são construções fundamentais utlizadas para organizar os dados de modo que possam ser acessados e usados.

|Set|Estrutura que permite adicionar e remover dados não duplicados|{ }|`{'C3', 25, 7.5}`|

## Aplicando uma função a cada elemento da sequência

```python
def calcular_horas(dias):
    return 24 * dias
```

```python
dias = [1, 10, 15, 365, 720]
```

```python
resultado = []
for d in dias:
    r = calcular_horas(d)
    resultado.append(r)
```
