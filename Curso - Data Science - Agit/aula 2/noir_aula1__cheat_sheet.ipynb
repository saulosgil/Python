**Métodos de Strings**


  | |Método|Descrição|Exemplo - Entada| Exemplo - Saída |
  |--|------|---------|-------|----|
  |1|`.lower()`|retorna cópia com todas<br>letras convertidas em minúsculas| ```'Mathema Code'.lower()```|`'mathema code'`|
  |2|`.upper()`|retorna cópia com todas<br> letras convertidas em maiúsculas| ```'Mathema Code'.upper()``` |`'MATHEMA CODE'`|
  |3|`.replace()`|substitui uma <br>substring por outra| ```'Mathema CO$DE'.replace('CO$DE', 'Code'])``` | `Mathema Code`|
  |4|`.strip()`|remove espaços <br>no início e no fim| ```' Mathema Code  '.strip()``` | 'Mathema Code'|
  |5|`.capitalize()`|retorna cópia capitalizada| ```'mathema code'.capitalized()``` | `'Mathema Code'`|
  |6|`.find()`|retorna o menor índice <br>da substring encontrada (ou -1 se não encontrar)| ```'Mathema Code'.find('Code')``` | 9|  
  |7|`.rjust()`|ajusta comprimento com <br>caracteres desejados| ```'11'.rjust(3, '0')``` | '011'|
  |8|`.split()`|quebra a string em<br> várias palavras| ```'Mathema Code'.split()``` | `['Mathema', 'Code']`|
  |9|`.join()`|junta uma lista de <br>strings em uma única string| ```' '.split(['Mathema', 'Code'])``` | `'Mathema Code'`|
  |10|`.startswith()`|retorna **True** se a string<br>iniciar com string fornecida| ```'Mathema Code'.startswith('Math'])``` | `True`|
  |11|`.endswith()`|retorna True se string <br>iniciar com string fornecida| ```'Mathema Code'.endswith('Code'])``` | `True`|
