```python
split = nomes[0].split()
r = []
for n in split[1:-1]:
    if n[0] == n[0].upper():
        r.append(n[0])
    else:
        r.append(n)
split[0] + ' ' + ' '.join(r) + ' '  + split[-1]
```

```python
def abreviar_nome(nome:str)->str:
    split = nome.split()
    r = []
    for n in split[1:-1]:
        if n[0] == n[0].upper():
            r.append(n[0])
        else:
            r.append(n)
    return split[0] + ' ' + ' '.join(r) + ' '  + split[-1]
```

```python
abreviar_nome(nomes[0])
```

```python
r = []
for n in nomes:
    nome_abreviado = abreviar_nome(n) 
    r.append(nome_abreviado)
r    
```

```python
resposta = []
for nome in nomes:
    split = nome.split()
    r = []
    for n in split[1:-1]:        
        r.append(n[0])
        
    nome_abreviado = split[0] + ' ' + ' '.join(r) + ' '  + split[-1]
    resposta.append(nome_abreviado)
resposta
```
